```python
def filter_models(all_models):
    df_list = []
    for model in all_models:
        df_model = df[df['Name'].str.contains(model, case=False, na=False)]
        df_list.append(df_model)
    df_filtered = pd.concat(df_list, ignore_index=True)
    df_filtered = df_filtered[["Name", "Announced", "Display Type", "Display Size", "Resolution", "OS", "Chipset", "CPU", "Memory(Internal)", "Main Camera", "Front Camera", "NFC", "USB", "Sensors", "Battery Type", "Charging", "Colors"]]
    return df_filtered
```

```python
df_brand = filter_models(samsung_models)
df_brand
```

```python
df_vivo_transposed = df_brand.set_index("Name").transpose()
df_vivo_transposed
```
